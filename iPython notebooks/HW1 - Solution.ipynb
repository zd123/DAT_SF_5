## 1. How many passengers are in our passenger list?

## 2. What is the overall survival rate?

```python
data.Survived.head()
```

Since Survived is a binary variable, the survival rate corresponds to the average over the whole column.

$ mean = \frac{1}{N} \sum_{i = 0}^{N_{survived}}{S_i}  = \frac{N_{Survived}}{N}$

```python
survival_rate = data['Survived'].mean()
```

Since I'm asked for a * rate* I may prefer to format it as a percentage value

```python
print round(survival_rate*100,1), "%"
```
